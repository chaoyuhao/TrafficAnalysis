```python
# 初步尝试，获取一张图（未明确是哪张图，应该是第1张）

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from PIL import Image
from io import BytesIO
import requests

# 初始化 WebDriver
driver = webdriver.Chrome()  # 确保已安装 ChromeDriver
url = "https://www.weatherbug.com/traffic-cam/?latlng=34.193626,-77.909042&camId=11199"

try:
    # 打开目标页面
    driver.get(url)

    # 等待页面加载，最多等待 10 秒
    wait = WebDriverWait(driver, 10)

    # 定位实时交通图片的 <img> 标签，使用更具体的选择器
    traffic_image = wait.until(
        EC.presence_of_element_located((By.XPATH, "//img[contains(@src, 'traffic')]"))
    )

    # 获取实时交通图片的 URL
    image_url = traffic_image.get_attribute("src")
    print(f"实时交通图片 URL: {image_url}")

    # 请求并显示图片
    response = requests.get(image_url)
    if response.status_code == 200:
        image = Image.open(BytesIO(response.content))
        image.show()  # 显示图片
    else:
        print(f"无法加载图片，状态码: {response.status_code}")

except Exception as e:
    print(f"出现错误: {e}")

finally:
    driver.quit()  # 关闭浏览器

```

```python
# 初步尝试，监测一张图的更新时间（未明确是哪张图，应该是第1张）

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from PIL import Image
from io import BytesIO
import requests
import time
from datetime import datetime

# 初始化 WebDriver
driver = webdriver.Chrome()  # 确保已安装 ChromeDriver
url = "https://www.weatherbug.com/traffic-cam/?latlng=34.193626,-77.909042&camId=11199"

try:
    # 打开目标页面
    driver.get(url)
    print("正在加载页面，请稍候...")

    # 等待页面加载并获取初始图片
    wait = WebDriverWait(driver, 15)  # 最多等待 15 秒
    traffic_image = wait.until(
        EC.presence_of_element_located((By.XPATH, "//img[contains(@src, 'traffic')]"))
    )

    # 获取初始图片 URL
    previous_image_url = traffic_image.get_attribute("src")
    print(f"初始图片 URL: {previous_image_url}")

    start_time = time.time()  # 记录初始时间

    # 实时监测图片变化
    while True:
        # 获取当前图片 URL
        current_image_url = traffic_image.get_attribute("src")

        if current_image_url != previous_image_url:
            # 图片发生变化，记录时间
            change_time = datetime.now()
            elapsed_time = round(time.time() - start_time, 2)
            print(f"[{change_time}] 图片发生变化！距离上次变化间隔：{elapsed_time} 秒")

            # 更新图片 URL 和计时器
            previous_image_url = current_image_url
            start_time = time.time()

        else:
            print(f"[{datetime.now()}] 图片未发生变化")

        # 每隔 1 秒检查一次
        time.sleep(1)

except KeyboardInterrupt:
    print("\n监测已停止")
except Exception as e:
    print(f"出现错误: {e}")
finally:
    driver.quit()  # 关闭浏览器
```
